```python
import pandas as pd
import numpy as np

# Реальные средние температуры (примерные данные) для городов по сезонам
seasonal_temperatures = {
    "New York": {"winter": 0, "spring": 10, "summer": 25, "autumn": 15},
    "London": {"winter": 5, "spring": 11, "summer": 18, "autumn": 12},
    "Paris": {"winter": 4, "spring": 12, "summer": 20, "autumn": 13},
    "Tokyo": {"winter": 6, "spring": 15, "summer": 27, "autumn": 18},
    "Moscow": {"winter": -10, "spring": 5, "summer": 18, "autumn": 8},
    "Sydney": {"winter": 12, "spring": 18, "summer": 25, "autumn": 20},
    "Berlin": {"winter": 0, "spring": 10, "summer": 20, "autumn": 11},
    "Beijing": {"winter": -2, "spring": 13, "summer": 27, "autumn": 16},
    "Rio de Janeiro": {"winter": 20, "spring": 25, "summer": 30, "autumn": 25},
    "Dubai": {"winter": 20, "spring": 30, "summer": 40, "autumn": 30},
    "Los Angeles": {"winter": 15, "spring": 18, "summer": 25, "autumn": 20},
    "Singapore": {"winter": 27, "spring": 28, "summer": 28, "autumn": 27},
    "Mumbai": {"winter": 25, "spring": 30, "summer": 35, "autumn": 30},
    "Cairo": {"winter": 15, "spring": 25, "summer": 35, "autumn": 25},
    "Mexico City": {"winter": 12, "spring": 18, "summer": 20, "autumn": 15},
}

# Сопоставление месяцев с сезонами
month_to_season = {12: "winter", 1: "winter", 2: "winter",
                   3: "spring", 4: "spring", 5: "spring",
                   6: "summer", 7: "summer", 8: "summer",
                   9: "autumn", 10: "autumn", 11: "autumn"}

# Генерация данных о температуре
def generate_realistic_temperature_data(cities, num_years=10):
    dates = pd.date_range(start="2010-01-01", periods=365 * num_years, freq="D")
    data = []

    for city in cities:
        for date in dates:
            season = month_to_season[date.month]
            mean_temp = seasonal_temperatures[city][season]
            # Добавляем случайное отклонение
            temperature = np.random.normal(loc=mean_temp, scale=5)
            data.append({"city": city, "timestamp": date, "temperature": temperature})

    df = pd.DataFrame(data)
    df['season'] = df['timestamp'].dt.month.map(lambda x: month_to_season[x])
    return df

# Генерация данных
data = generate_realistic_temperature_data(list(seasonal_temperatures.keys()))
data.to_csv('temperature_data.csv', index=False)
```

##Код для анализа с распараллеливанием и без распараллеливания:

```python
from multiprocessing import Pool
import time
import matplotlib.pyplot as plt
from functools import partial

def analyze_city(city_data):
    """
    Анализ данных для одного города
    """
    # Скользящее среднее с min_periods=1
    city_data['rolling_mean'] = city_data['temperature'].rolling(
        window=30, min_periods=1
    ).mean()

    # Скользящее стандартное отклонение с min_periods=2
    city_data['rolling_std'] = city_data['temperature'].rolling(
        window=30, min_periods=2
    ).std()

    # Заполнение NaN в rolling_std используя bfill()
    city_data['rolling_std'] = city_data['rolling_std'].bfill()

    # Рассчет сезонной статистики
    seasonal_stats = city_data.groupby('season').agg({
        'temperature': ['mean', 'std']
    })

    # Преобразование индексов и названий колонок
    seasonal_stats.columns = ['seasonal_mean', 'seasonal_std']
    seasonal_stats = seasonal_stats.reset_index()

    # Присоединение сезонной статистики
    city_data = city_data.merge(seasonal_stats, on='season')

    # Выявление аномалий (используем оба метода)
    city_data['rolling_anomaly'] = np.abs(
        city_data['temperature'] - city_data['rolling_mean']
    ) > 2 * city_data['rolling_std']

    city_data['seasonal_anomaly'] = np.abs(
        city_data['temperature'] - city_data['seasonal_mean']
    ) > 2 * city_data['seasonal_std']

    return city_data

def analyze_data_parallel(data, n_processes=None):
    """
    Параллельный анализ данных
    """
    cities = data['city'].unique()
    if n_processes is None:
        n_processes = min(len(cities), 4)  # Ограничиваем максимальное число процессов

    with Pool(processes=n_processes) as pool:
        results = pool.map(
            analyze_city,
            [data[data['city'] == city].copy() for city in cities]
        )
    return pd.concat(results)

def analyze_data_sequential(data):
    """
    Последовательный анализ данных
    """
    return pd.concat([
        analyze_city(data[data['city'] == city].copy())
        for city in data['city'].unique()
    ])

def plot_city_analysis(data, city):
    """
    Визуализация результатов анализа для конкретного города
    """
    city_data = data[data['city'] == city]

    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 12))

    # График скользящих значений
    ax1.plot(city_data['timestamp'], city_data['temperature'],
             label='Температура', alpha=0.7)
    ax1.plot(city_data['timestamp'], city_data['rolling_mean'],
             label='Скользящее среднее', color='orange')
    ax1.scatter(city_data[city_data['rolling_anomaly']]['timestamp'],
                city_data[city_data['rolling_anomaly']]['temperature'],
                color='red', label='Аномалии (скользящие)', alpha=0.6)
    ax1.set_title(f'Анализ температуры для {city} (скользящие значения)')
    ax1.set_xlabel('Дата')
    ax1.set_ylabel('Температура (°C)')
    ax1.grid(True, linestyle='--', alpha=0.7)
    ax1.legend()

    # График сезонных значений
    ax2.plot(city_data['timestamp'], city_data['temperature'],
             label='Температура', alpha=0.7)
    ax2.plot(city_data['timestamp'], city_data['seasonal_mean'],
             label='Сезонное среднее', color='green')
    ax2.scatter(city_data[city_data['seasonal_anomaly']]['timestamp'],
                city_data[city_data['seasonal_anomaly']]['temperature'],
                color='purple', label='Аномалии (сезонные)', alpha=0.6)
    ax2.set_title(f'Анализ температуры для {city} (сезонные значения)')
    ax2.set_xlabel('Дата')
    ax2.set_ylabel('Температура (°C)')
    ax2.grid(True, linestyle='--', alpha=0.7)
    ax2.legend()

    plt.tight_layout()
    plt.show()

# Пример использования:
if __name__ == '__main__':
    # Загрузка данных
    data = pd.read_csv('temperature_data.csv', parse_dates=['timestamp'])

    # Измерение времени выполнения для последовательного анализа
    start_time = time.time()
    sequential_result = analyze_data_sequential(data)
    sequential_time = time.time() - start_time

    # Измерение времени выполнения для параллельного анализа
    start_time = time.time()
    parallel_result = analyze_data_parallel(data)
    parallel_time = time.time() - start_time

    # Вывод времени выполнения
    print(f"Время выполнения без распараллеливания: {sequential_time:.2f} секунд")
    print(f"Время выполнения с распараллеливанием: {parallel_time:.2f} секунд")

    # Визуализация результатов для Москвы
    plot_city_analysis(parallel_result, 'Moscow')
```

#Возможные причины:
- Небольшой объем данных: Если объем данных относительно мал, накладные расходы на создание и управление потоками могут превышать выигрыш от параллельного выполнения. Это особенно актуально для задач, которые быстро выполняются в последовательном режиме.

- Затраты на межпроцессное взаимодействие: Распараллеливание требует передачи данных между процессами. Это может быть дорогостоящим в плане времени, особенно если данные большие или если выполняется большое количество мелких задач.

- Использование нескольких ядер: Если ваш процессор имеет ограниченное количество ядер, распараллеливание может не дать прироста производительности, так как все процессы будут конкурировать за одни и те же ресурсы.

- Балансировка нагрузки: Если данные по городам сильно различаются по размеру, одни процессы могут закончить работу значительно быстрее других, что приведет к неэффективному использованию ресурсов.

- IO-задержки: Если чтение и запись данных включают значительные задержки ввода-вывода, они могут замедлить процесс параллельного выполнения, так как все процессы будут ждать завершения операций чтения/записи.

##Мониторинг текущей температуры:

```python
import aiohttp
import asyncio
import pandas as pd
from datetime import datetime
import time
import requests

OWM_API_KEY = "API"

async def fetch_temperature(session, city):
    url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&units=metric&APPID={OWM_API_KEY}"
    async with session.get(url) as response:
        data = await response.json()
        if 'main' in data and 'temp' in data['main']:
            return city, data['main']['temp']
        else:
            return city, None

def fetch_temperature_sync(city):
    url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&units=metric&APPID={OWM_API_KEY}"
    response = requests.get(url)
    data = response.json()
    if 'main' in data and 'temp' in data['main']:
        return city, data['main']['temp']
    else:
        return city, None

def get_current_season():
    month = datetime.now().month
    if month in [12, 1, 2]:
        return "winter"
    elif month in [3, 4, 5]:
        return "spring"
    elif month in [6, 7, 8]:
        return "summer"
    else:
        return "autumn"

async def main_async(cities):
    async with aiohttp.ClientSession() as session:
        tasks = [fetch_temperature(session, city) for city in cities]
        results = await asyncio.gather(*tasks)
    return dict(results)

def main_sync(cities):
    results = {}
    for city in cities:
        city, temp = fetch_temperature_sync(city)
        results[city] = temp
    return results

def analyze_temperatures(current_temperatures, historical_data, current_season):
    for city, temp in current_temperatures.items():
        if temp is not None:
            city_data = historical_data[historical_data['city'] == city]
            season_data = city_data[city_data['season'] == current_season]

            if not season_data.empty:
                mean_temp = season_data['temperature'].mean()
                std_temp = season_data['temperature'].std()

                if mean_temp - 2 * std_temp <= temp <= mean_temp + 2 * std_temp:
                    print(f"Температура в {city} ({temp}°C) в норме для сезона.")
                else:
                    print(f"Температура в {city} ({temp}°C) является аномальной для сезона.")
            else:
                print(f"Нет данных для сезона {current_season} в {city}.")
        else:
            print(f"Не удалось получить данные о температуре для {city}.")

# Загрузка исторических данных
historical_data = pd.read_csv('temperature_data.csv', parse_dates=['timestamp'])
cities = ["Berlin", "Cairo", "Dubai", "Beijing", "Moscow"]
current_season = get_current_season()

# Асинхронное выполнение
start_time = time.time()
current_temperatures_async = await main_async(cities)
async_duration = time.time() - start_time

# Синхронное выполнение
start_time = time.time()
current_temperatures_sync = main_sync(cities)
sync_duration = time.time() - start_time

# Анализ температур
print("\nАсинхронный подход:")
analyze_temperatures(current_temperatures_async, historical_data, current_season)

print("\nСинхронный подход:")
analyze_temperatures(current_temperatures_sync, historical_data, current_season)

# Вывод времени выполнения
print(f"\nВремя выполнения асинхронного подхода: {async_duration:.2f} секунд")
print(f"Время выполнения синхронного подхода: {sync_duration:.2f} секунд")
```

##Вывод: в данном случае асинхронный подход, показывает себя лучше.
Несмотря на то, что данные были изменены специально, в Пекине и Москве температура указана, как нормальная, возможно для этих городов погода аномальная. Необходимо проверить в другой день.
